```python
import subprocess

def download_aria2c(url, output, dest):
    cmd = [
        "aria2c", "-x", "16", "-s", "16", "-k", "1M",
        "-d", dest,
        "-o", output,
        url
    ]
    print(f"⬇️ Descargando: {output}")
    result = subprocess.run(cmd)
    if result.returncode != 0:
        print(f"❌ Error al descargar {output}")
    else:
        print(f"✅ Descargado: {output}")

# Directorio base
base_dir = "/content/drive/MyDrive/ComfyUI"

# Modelos base
download_aria2c(
    "https://civitai.com/api/download/models/170706",
    "RealMixEuropean1.safetensors",
    f"{base_dir}/models/checkpoints"
)

download_aria2c(
    "https://civitai.com/api/download/models/241394",
    "cyberrealisticPony_v8.safetensors",
    f"{base_dir}/models/checkpoints/Pony Realistic"
)

# VAEs
download_aria2c(
    "https://huggingface.co/stabilityai/sd-vae-ft-mse-original/resolve/main/vae-ft-mse-840000-ema-pruned.safetensors",
    "vae-ft-mse-840000-ema-pruned.safetensors",
    f"{base_dir}/models/vae"
)

download_aria2c(
    "https://huggingface.co/madebyollin/sdxl-vae-fp16-fix/resolve/main/sdxl_vae.safetensors",
    "sdxl_vae.safetensors",
    f"{base_dir}/models/vae"
)

# LoRAs
download_aria2c(
    "https://civitai.com/api/download/models/106469",
    "DetailedEyes_V3.safetensors",
    f"{base_dir}/models/lora"
)

download_aria2c(
    "https://civitai.com/api/download/models/144260",
    "hand_5.5-SDXL.safetensors",
    f"{base_dir}/models/lora"
)

download_aria2c(
    "https://civitai.com/api/download/models/152036",
    "fractalism_last.safetensors",
    f"{base_dir}/models/lora"
)

# Upscalers
download_aria2c(
    "https://huggingface.co/uwg/upscaler/resolve/main/4x-UltraSharp.pth",
    "4x-UltraSharp.pth",
    f"{base_dir}/models/upscale_models"
)

download_aria2c(
    "https://huggingface.co/uwg/upscaler/resolve/main/4x_NMKD-Siax_200k.pth",
    "4x_NMKD-Siax_200k.pth",
    f"{base_dir}/models/upscale_models"
)

# Embeddings
download_aria2c(
    "https://huggingface.co/datasets/Nerfgun3/bad_prompt/resolve/main/bad_prompt.pt",
    "bad_prompt.pt",
    f"{base_dir}/embeddings"
)

download_aria2c(
    "https://huggingface.co/datasets/Nerfgun3/bad_prompt/resolve/main/bad_prompt_version2.pt",
    "bad_prompt_version2.pt",
    f"{base_dir}/embeddings"
)

download_aria2c(
    "https://huggingface.co/datasets/Nerfgun3/badhandv4/resolve/main/badhandv4.pt",
    "badhandv4.pt",
    f"{base_dir}/embeddings"
)

# YOLO face detector
download_aria2c(
    "https://huggingface.co/keremberke/yolov8m-face/resolve/main/yolov8m-face.pt",
    "face_yolov8m.pt",
    f"{base_dir}/models/ultralytics"
)
```

### Run ComfyUI with colab iframe (use only in case the previous way with localtunnel doesn't work)

You should see the ui appear in an iframe. If you get a 403 error, it's your firefox settings or an extension that's messing things up.

If you want to open it in another window use the link.

Note that some UI features like live image previews won't work because the colab iframe blocks websockets.

```python
import threading
import time
import socket
def iframe_thread(port):
  while True:
      time.sleep(0.5)
      sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
      result = sock.connect_ex(('127.0.0.1', port))
      if result == 0:
        break
      sock.close()
  from google.colab import output
  output.serve_kernel_port_as_iframe(port, height=1024)
  print("to open it in a window you can open this link here:")
  output.serve_kernel_port_as_window(port)

threading.Thread(target=iframe_thread, daemon=True, args=(8188,)).start()

!python main.py --dont-print-server
```
